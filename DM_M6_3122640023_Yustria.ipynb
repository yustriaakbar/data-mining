```python
# Impor pustaka
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import graphviz
from sklearn.tree import export_text

# 1. Tampilkan dataset train titanic.csv
train_data = pd.read_csv('titanic.csv')
# print(train_data.head())

# 2. Tampilkan dataset titanic_test.csv
test_data = pd.read_csv('titanic_test.csv')
# print(test_data.head())

# 3. Isi missing value pada fitur Age dengan nilai mean dari masing-masing class
age_mean_by_class = train_data.groupby('Pclass')['Age'].mean()
train_data['Age'] = train_data.apply(lambda row: age_mean_by_class[row['Pclass']]
                                     if pd.isnull(row['Age']) else row['Age'], axis=1)
test_data['Age'] = test_data.apply(lambda row: age_mean_by_class[row['Pclass']]
                                   if pd.isnull(row['Age']) else row['Age'], axis=1)
# Menampilkan baris-baris yang telah diisi nilai Age
# with pd.option_context('display.max_rows', None):
#     print(train_data[['Pclass', 'Age']])

# 4. Ambil kolom fitur (Sex, Age, Pclass, Fare) dari train_data dan test_data
features = ['Sex', 'Age', 'Pclass', 'Fare']
X_train = train_data[features]
X_test = test_data[features]

# Encode kolom 'Sex' menjadi numerik
label_encoder = LabelEncoder()
X_train['Sex'] = label_encoder.fit_transform(X_train['Sex'])
X_test['Sex'] = label_encoder.transform(X_test['Sex'])

# 5. Ambil kolom kelas (Survived) dari train_data
train_label = train_data['Survived']

# Latih model Decision Tree dan hitung error rate
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train, train_label, test_size=0.2, random_state=42)
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train_split, y_train_split)
predictions = decision_tree.predict(X_val_split)
error_rate = 1 - accuracy_score(y_val_split, predictions)

# Tampilkan hirarki dari Decision Tree
tree_rules = export_text(decision_tree, feature_names=features)
print("Decision Tree Rules:")
print(tree_rules)

# Tampilkan error rate
print("Error Rate:", error_rate)
```
